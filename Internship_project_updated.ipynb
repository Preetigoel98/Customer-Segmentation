```python
from datetime import datetime 

# Step 1: Load the cleaned dataset from previous step
df = pd.read_csv("marketing_campaign_cleaned.csv")

# Step 2: Create 'Age' column
current_year = datetime.now().year
df["Age"] = current_year - df["Year_Birth"]

# Step 3: Create 'Customer_Tenure' in months
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], errors="coerce")
df["Customer_Tenure_Months"] = (datetime.now() - df["Dt_Customer"]).dt.days // 30

# Step 4: Create 'Children' (Kidhome + Teenhome)
df["Children"] = df["Kidhome"] + df["Teenhome"]

# Step 5: Create 'Total_Spending' (sum of all product spending columns)
spending_cols = [
    "MntWines", "MntFruits", "MntMeatProducts", 
    "MntFishProducts", "MntSweetProducts", "MntGoldProds"
]
df["Total_Spending"] = df[spending_cols].sum(axis=1)

# Step 6: Create 'Total_Purchases' (sum of all purchase channels)
purchase_cols = [
    "NumDealsPurchases", "NumWebPurchases", 
    "NumCatalogPurchases", "NumStorePurchases"
]
df["Total_Purchases"] = df[purchase_cols].sum(axis=1)

# Step 7: Verify new columns
print("\n----- New Columns Added -----")
print(df[["Age", "Customer_Tenure_Months", "Children", "Total_Spending", "Total_Purchases"]].head())

# Step 8: Save engineered dataset
df.to_csv("marketing_campaign_featured.csv", index=False)
print("\n✅ Feature Engineering done! Saved as 'marketing_campaign_featured.csv'")
```

```python
#DBSCAN (Density-Based Clustering)

from sklearn.cluster import DBSCAN

# DBSCAN clustering
dbscan = DBSCAN(eps=2, min_samples=5)  # eps and min_samples need tuning
labels_dbscan = dbscan.fit_predict(df_pca)

# Add cluster labels
df['Cluster_DBSCAN'] = labels_dbscan
df[['Cluster_DBSCAN']].head()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

eps_values = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]
min_samples_values = [3, 5, 7, 10]

best_score = -1
best_params = {}

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(df_pca)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        
        # Skip if all points are noise or only 1 cluster
        if n_clusters < 2:
            continue
        
        # Calculate silhouette score excluding noise
        mask = labels != -1
        score = silhouette_score(df_pca[mask], labels[mask])
        
        print(f"eps={eps}, min_samples={min_samples}, clusters={n_clusters}, silhouette={score:.3f}")
        
        if score > best_score:
            best_score = score
            best_params = {'eps': eps, 'min_samples': min_samples, 'n_clusters': n_clusters}

print("Best DBSCAN parameters:", best_params)
print("Best Silhouette Score:", best_score)

```

Observation from your results
eps = 0.5, min_samples = 3 → silhouette = 0.898 ✅ but most points are labeled as noise.
eps > 1.5 → silhouette drops significantly (0.1–0.2) but clusters cover more points.
eps = 4.5, min_samples = 5 → silhouette = 0.472, only 3 clusters → still better coverage than eps=0.5.

Key point:
DBSCAN’s silhouette score is misleading when most points are -1 (noise).
The best “usable” DBSCAN clustering is not the highest silhouette, but the one that gives:
Most points assigned to clusters (not noise)
Reasonable number of clusters for profiling
